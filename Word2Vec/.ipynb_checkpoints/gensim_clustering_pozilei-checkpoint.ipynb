```python
import numpy as np
from itertools import chain
import gensim
import scipy
import pandas as pd
import logging
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
from matplotlib import pyplot
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_json('processed_data/preprocessed_data.json', orient='columns', encoding='utf-8')
```

```python
data.head()
```

```python
# from tensorflow notebook:
#combine headlines with text so we have one column to be analyzed:
def combineTextToBeAnalyzed(processed_headline, processed_article):
    """
    So this one is a bit dirty: I figured out, that some of the headlines have more than one sentence (mostly 2).
    Instead of a list with the words of one sentence, you get a list of a list (sentencens -> words).
    I want the Headlines and the articles to be combined, beeing the article the leading part of the combination.
    Since i'm not sure how to explicitely accomplish that task in Pandas, i broke down the Problem to a List-Merge.
    Es gibt noch ein kleines Problem, wenn man diese Funktion hier mehrmals ausführt, multiplizieren sich die Headlines.
    Das ist komisch und liegt vermutlich an der Referenzierung der dataframes. Bitte immer die Daten neu einladen,
    bevor man das hier ausführt (Eine Zeile davor)
    """
    list_of_headlines = data["processed_headlines"].tolist()
    list_of_articles = data["processed_articles"].tolist()
    combined_list = []
    # In the List of Articles of Sentences, Insert the Headline before the first sentence
    for i in range(len(list_of_headlines)):
        #If the Headline consists of more than one sentence, append to list of articles
        if any(isinstance(el, list) for el in list_of_headlines[i]):
            list_of_articles[i] = list_of_headlines[i] + list_of_articles[i]
        #If the Headline is only one sentence, insert to articles
        else:
            list_of_articles[i].insert(0, list_of_headlines[i])
    return list_of_articles
```

```python
data['combinedText'] = combineTextToBeAnalyzed(data['processed_headlines'], data['processed_articles'])
```

```python
sentences = data['combinedText'].tolist()
```

```python
len(sentences)
```

```python
# flatten to get only sentences (documents level being discarded)

sentences_flattened = list(chain.from_iterable(sentences))
```

```python
sentences_flattened[0:7]
```

### Model
- size: (default 100) The number of dimensions of the embedding, e.g. the length of the dense vector to represent each token (word).
- window: (default 5) The maximum distance between a target word and words around the target word.
- min_count: (default 5) The minimum count of words to consider when training the model; words with an occurrence less than this count will be ignored.
- workers: (default 3) The number of threads to use while training.
- sg: (default 0 or CBOW) The training algorithm, either CBOW (0) or skip gram (1).

```python
# config logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
# train model with default params
model = Word2Vec(sentences_flattened)
```

```python
# print the learned vocabulary of tokens
words = list(model.wv.vocab)
print(words[0:20])
```

```python
# review the embedded vector for a specific token
print(model['verkehrsunfall'])
```

```python
# plot tokens (first try)
# fit a 2d PCA model to the vectors
X = model[model.wv.vocab]
pca = PCA(n_components=2)
result = pca.fit_transform(X)
# create a scatter plot of the projection
pyplot.scatter(result[:, 0], result[:, 1])
words = list(model.wv.vocab)
for i, word in enumerate(words):
    pyplot.annotate(word, xy=(result[i, 0], result[i, 1]))
pyplot.show()
```

## TSNE
### "How to Use t-SNE Effectively
Although extremely useful for visualizing high-dimensional data, t-SNE plots can sometimes be mysterious or misleading. By exploring how it behaves in simple cases, we can learn to use it more effectively." https://distill.pub/2016/misread-tsne/

```python
# plot with tsne
# play with perplexity! (between 1 and 50)

def tsne_plot(model):
    "Creates a TSNE model and plots it"
    labels = []
    tokens = []

    for word in model.wv.vocab:
        tokens.append(model[word])
        labels.append(word)
    
    tsne_model = TSNE(perplexity=3, n_components=2, init='pca', n_iter=2500, random_state=23)
    new_values = tsne_model.fit_transform(tokens)

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(16, 16)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')
    plt.show()
```

```python
tsne_plot(model)
```

```python
# second model where only words occurring at least 300 times are considered
model2 = Word2Vec(sentences_flattened, min_count=300)
```

```python
# plot for second model
tsne_plot(model2)
```
